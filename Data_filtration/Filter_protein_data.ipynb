```python
###################
#Lenovo L13       #
#python version3  #
#Mohamed Emam     #
#Protein filter   #
###################
#Import library
import pandas as pd
#Read the file
protein= pd.read_csv("/Users/mmansour/Desktop/raw_protein_data.csv")
#Add sum coloumn total
protein['Total'] = protein.sum(axis=1)

#Extract protein with zero value across the samples
remove_features = protein.loc[protein['Total'] == 0]
remove_features
```

```python
#Filteration_of_removed_features_index
protein_features_index = remove_features.index
index_list = list(protein_features_index)
protein = protein.drop(protein.index[index_list])
protein= protein.iloc[0:, 0:(len(protein.columns)-1)]
#Export_csv
protein.to_csv("/Users/mmansour/Desktop/MS_filtered.csv", index=False)
```
